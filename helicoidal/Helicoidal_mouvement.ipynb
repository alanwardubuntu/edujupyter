# Mouvement hélicoïdale

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

t = np.arange(0.0, 3.0, 0.05)
r = 1.0 
vh = 4.0
vv = 0.1
omega = vh / r

x = r * np.cos(omega * t)
y = r * np.sin(omega * t)
z = vv * t

fig = plt.figure()
ax = fig.gca(projection='3d')

ax.plot(x, y, linestyle="dashed", color="grey", zs=0.0)
ax.plot(x, y, z, linestyle="solid", color="blue", linewidth=2)
plt.ioff()
```
