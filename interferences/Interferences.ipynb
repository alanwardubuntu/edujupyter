# Interférences entre deux fronts d'ondes cohérentes


Une lumière cohérente (= mêmes amplitude, fréquence et vitesse de propagation) 
est émise en direction de deux fentes. Ces fentes sont situées aux coordonnées 
(0, 0.5, 0) et (0, -0.5, 0), et supposées étroites en relation avec la longueur 
d'onde de la lumière. La source de la lumière est située au loin su l'axe Ox.

### Distances à partir des deux fentes

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  
import numpy as np

x = np.arange(0.0, 2.0, 0.01)
y = np.arange(-1.0, 1.0, 0.01)

d1 = np.zeros((len(x),len(y)))
d2 = np.zeros((len(x),len(y)))
for i in range(0, len(x)):
    for j in range(0, len(y)):
        d1[i][j] = np.sqrt(x[i]*x[i] + (y[j] - 0.5)*(y[j] - 0.5))
        d2[i][j] = np.sqrt(x[i]*x[i] + (y[j] + 0.5)*(y[j] + 0.5))

X, Y = np.meshgrid(x, y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_title("Distances a partir des fentes")
ax.set_zlim([-2.0, 3.0])

ax.plot_surface(X, Y, d1, alpha=0.5)
ax.plot_surface(X, Y, d2, alpha=0.5)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

### Équations des ondes

```python
k = 2 * np.pi / 0.25 # longeur d'onde (m)
phi1 = np.zeros((len(x),len(y)))
phi2 = np.zeros((len(x),len(y)))
for i in range(0, len(x)):
    for j in range(0, len(y)):
        phi1[i][j] = np.cos(-k * d1[i][j])
        phi2[i][j] = np.cos(-k * d2[i][j])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_title("Les deux ondes")
ax.set_zlim([-2.0, 2.0])

ax.plot_surface(X, Y, phi1, alpha=0.35)
ax.plot_surface(X, Y, phi2, alpha=0.35)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

### Superposition (somme) des deux ondes

```python

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_zlim([-5.0, 5.0])

phi = np.zeros((len(x),len(y)))
phi = phi1 + phi2

ax.plot_surface(X, Y, phi, color='g', alpha=0.5)
ax.set_title("La somme des deux ondes")
ax.set_zlim([-2.0, 2.0])

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

## Déphasage entre les deux ondes

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_zlim([-5.0, 5.0])

dp = np.zeros((len(x),len(y)))
dp = np.cos(k * np.abs(d1 - d2))

ax.plot_surface(X, Y, dp, alpha=0.5)
ax.set_title("Dephasage entre les deux ondes")
ax.set_zlim([-1.2, 1.2])

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
