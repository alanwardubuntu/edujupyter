# Dérivation de fonctions

Començons per définir une function simple:

    f: R -> R
       x -> f(x) = x^2 - 2 * x * sin(x)



Dériver cette fonction *par rapport à x* équivaut à calculer sa pente par rapport à x: pour une petite variation dx donnée, quelle sera la variation df correspondante?  

Començons par définir notre function de sortie:

```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt


def f(x):
    return (x**2 - 2 * x * sin(x))
```

Représentons la fonction sur une intervalle. Nous pouvons y observer des zones de croissance, et d'autres de décroissance:

```python
x = np.arange(-1.0, 9.0, 0.1)
y = []
for i in range(0, len(x)):
    y.append(f(x[i]))

plt.plot (x, y)
```

En fixant un valeur très petit pour dx, par exemple 10^-5, nous pouvons obtenir **une approximation** de la dérivée de f, soit df / dx. Remarquons que cette dérivée est *positive* là ou la fonction originelle est croissante, et *négative* là où la fonction est décroissante:

```python
dx = 10**(-5)

y2 = []
for i in range(0, len(x)):
    df = f(x[i] + dx) - f(x[i])
    y2.append(df / dx)

plt.plot (x, y2, 'r')
```

Cette méthode d'approximation numérique sera souvent celle choisie par les ingénieurs, quand connaître l'expression exacte n'est pas une priorité. Par contre, pour les investigateurs en mathématiques ou en physique cette expression analytique sera nécessaire. Pour ce faire, nous pouvons utiliser les connaissances acquises en cours de mathématiques -- et que nous devons savoir employer. Alternativement, les systemes d'algèbre per ordinateur peuvent, aujourd'hui, nous aider pour vérifier ou compléter nos calculs:

```python

x = symbols('x')
f = x*x - 2*x*sin(x)

d = diff(f, x)
print ("f = %s" % f)
print ("df/dx = %s" % d)

p1 = plot(f, xlim=[-1, 9], ylim=[-5, 70])
p2 = plot(d, xlim=[-1, 9], ylim=[-5, 30])
```
