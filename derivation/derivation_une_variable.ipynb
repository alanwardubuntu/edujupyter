# Dérivation d'une fonction d'une variable

Commençons par prendre une fonction d'une seule variable, par exemple $f(x) = \sqrt{1 + x^2}$. Nous pouvons représenter la fonction:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-4.0, 4.0, 0.01)
f = np.sqrt(1.0 + x*x)

plt.plot(x, f)
plt.ioff()
```

Il est clair que cette fonction présente des valeurs de $x$ pour lesquelles elle est croissante, et d'autres pour lesquelles elle décroit. Plus exactement, nous pouvons calculer la _pente_ de la fonction pour tout valeur de $x$. Précisément, cette pente est exprimée par:

$ f'(x) = lim_{dx \rightarrow 0} \frac{f(x + dx) - f(x)}{dx} $

Cette nouvelle fonction es appellée _dérivée de f par rapport à x_. Quelques exemples de dérivée:

$3x$ -> $3$

$3x + 2$ -> $3$

$3x + x^2$ -> $3 + 2x$

$1 - 2x + 3x^2$ -> $-2 + 6x$

$ \frac{1}{x}$ -> $\frac{-1}{x^2}$

$\sqrt{x}$ -> $\frac{1}{2 \sqrt{x}}$

Vous allez étudier comment calculer ces expressions exactes en cours de Mathématiques. Ici, en Physique, nous aurons besoin de ces connaissances!

Par ailleurs, il est parfois difficile ou impossible de calcular l'expression exacte. Alors, il est souvent suffisant d'en prendre une _approximation numérique_. Si nous prenons une valeur assez petite de $dx$, cette approximation serà assez proche du valeur réel ... en tout cas, suffisamment pour que nous puissions faire une réflexion sur le problème physique.

```python
dx = 1.0e-5

fp = np.zeros(len(x))
for i in range(0, len(x)):
    xx = x[i]
    y1 = np.sqrt(1.0 + xx*xx)
    y2 = np.sqrt(1.0 + (xx + dx)*(xx + dx))
    fp[i] = (y2 - y1) / dx

    
plt.plot(x, f, label="f(x)")
plt.plot(x, fp, label="f'(x)")
plt.legend(loc='upper center')
plt.ioff()
```

Nous pouvons voir que pour la partie gauche du domaine, la fonction $f$ est décroissante et sa dérivée $f'$ est négative. Par contre, dans la partie droite, la fonction est croissante et sa dérivée est positive.

En Physique, nous aurons souvent tendance à dériver _par rapport au temps $t$_. 
Par exemple, soit un objet que a une trajectoire

$ x(t) = x_0 + v_0 . t - \frac{g.t^2}{2} $

La _dérivée de x par rapport à t_ est souvent notée $x'(t)$ ou ${\partial x} \over {\partial t}$ peut être calculée en appliquant les règles de calcul des dérivées. 
Quand nous dérivons une première fois par rapport au temps, nous obtenons la vitesse 
de l'objet:

$ v(t) = x'(t) = \frac{\partial x}{\partial t} = v_0 - g.t $

Si nous dérivons une deuxième fois par rapport au temps, nous obtenons son 
accélération:

$ a(t) = v'(t) = x"(t) = \frac{\partial v}{\partial t} = \frac{\partial^2 x}{\partial t^2} = -g$

Représentons les trois courbes:


```python
t = np.arange(0.0, 4.0, 0.1)
x0 = 0.0 # m, objet inicialement sur le sol
v0 = 15.0 # m/s, objet inicialement lancée vers le haut
g = 9.81 # m/s2, un vieux ami

a = -g + 0*t
v = v0 - g*t
x = x0 + v0*t - g*t*t/2

fig, ax = plt.subplots()
plt.plot(t, a, label="a(t)")
plt.plot(t, v, label="v(t)")
plt.plot(t, x, label="x(t)")
plt.legend(loc='lower center')

ax.annotate('vitesse nulle', xy=(1.5, 0.0), xytext=(1.7,0.1),
            arrowprops={'arrowstyle': '->'})
ax.annotate('hauteur maximale', xy=(1.5, 12.5), xytext=(1.2, 15.0),
            arrowprops={'arrowstyle': '->'})
ax.annotate('tombe par terre', xy=(3.0, 0.0), xytext=(3.0,5.0),
            arrowprops={'arrowstyle': '->'})

plt.ioff()
```
